```python
API_KEY = '2ab9accb3aa5b4067c0d16feb3c9afa6'

def get_lego_set_details(set_num):
    url = f'https://rebrickable.com/api/v3/lego/sets/{set_num}/'
    headers = {'Authorization': f'key {API_KEY}'}
    try:
        response = requests.get(url, headers=headers, timeout=10)  # timeout in seconds
        response.raise_for_status()  # Raise error for bad responses (4xx/5xx)
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Error for set {set_num}: {e}")
        return None
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

set_num = "77053-1"
url = f"https://www.brickeconomy.com/set/{set_num}/"

# Setup Chrome options
options = Options()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")

# Start driver
driver = webdriver.Chrome(options=options)
driver.get(url)

# Wait up to 10 seconds for page elements to appear
wait = WebDriverWait(driver, 10)

data = {}

try:
    # 🟢 Wait for and extract Estimated Value (usually in a box with $)
    estimate_box = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.set-value-box")))
    data["Estimated Value"] = estimate_box.text.strip()

    # 🟢 Get Retail Price (MSRP)
    msrp_label = wait.until(EC.presence_of_element_located((By.XPATH, "//div[text()='MSRP']")))
    msrp_value = msrp_label.find_element(By.XPATH, "following-sibling::div")
    data["Retail Price (MSRP)"] = msrp_value.text.strip()

    # 🟢 Get Year Released
    year_label = wait.until(EC.presence_of_element_located((By.XPATH, "//div[text()='Year Released']")))
    year_value = year_label.find_element(By.XPATH, "following-sibling::div")
    data["Year Released"] = year_value.text.strip()

    print(f"\n📦 Data for LEGO set {set_num}")
    for k, v in data.items():
        print(f"{k}: {v}")

except Exception as e:
    print(f"❌ Error scraping set {set_num}: {e}")

driver.quit()
```
