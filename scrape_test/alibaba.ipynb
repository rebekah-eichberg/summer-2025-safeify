```python
import undetected_chromedriver as uc
driver = uc.Chrome(headless=True,use_subprocess=False)
def get_page(URL):
    driver.get(URL)
    return BeautifulSoup(driver.page_source, "html.parser")
```

```python
def search_url(term:str,pageno:int):
    term=term.replace(' ','+')
    return "https://www.alibaba.com/trade/search?SearchText="+term+"&page="+str(pageno)+"&productAuthTag=CE%2CCPC%2CUKCA%2CGCC%2CEN+71"
```

```python
def get_item_features(item):
    image='https://'+item.find("img").attrs['src'][2:]
    title=item.find("a",attrs={"data-spm":"d_title"}).text
    link='https://'+item.find("a",attrs={"data-spm":"d_title"}).attrs['href'][2:]
    company_name=item.find("a",attrs={"data-spm":"d_companyName"}).text
    raw_certs=item.find_all("div",class_="search-card-e-icon__certification-wrapper")
    certs=[x.find("img").attrs['alt'] for x in raw_certs]
    return {"Title":title,"Image":image,"Company":company_name,"Link":link,"Certificates":certs}
```

```python
def convert_to_df(data):
    all_keys=set()
    for item in data:
        all_keys=all_keys.union(item.keys())
    data_dict=dict()
    for key in all_keys:
        data_dict[key]=list()
    for item in data:
        for key in all_keys:
            data_dict[key].append(item.get(key,''))
    return pd.DataFrame.from_dict(data_dict)
```

```python
def get_items_from_search(term:str,num_pages:int):
    item_data=list()
    for i in range(1,num_pages+1):
        page=get_page(search_url(term,i))
        raw_items=page.find_all("div",class_="fy23-search-card m-gallery-product-item-v2 J-search-card-wrapper fy23-gallery-card searchx-offer-item")
        if len(raw_items)==0:
            print("Error. No items found on page",i)
        else:
            print("Processing items for page", i)
            for item in raw_items:
                item_data.append(get_item_features(item))
    return convert_to_df(item_data)
```
